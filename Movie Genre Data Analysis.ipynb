# Movie Genre Data Analysis

### -- Loading Library 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

### Importing Data

```python
movies = pd.read_csv(r'C:\Users\DELL\Desktop\Projects\Python\Movie Genres Data Analysis\imdb_movies.csv')
pd.set_option('display.max.rows', 10966)
pd.set_option('display.max.columns', 25)
```

```python
movies.info()
```

### Cleaning Data
1. Check and Remove Duplicates

```python
movies[movies.duplicated()]
movies.drop_duplicates(inplace = True)
```

##### 2. Removing Null Data 

```python
movies.dropna(subset = ['genres'], inplace = True)
```

##### 3. Creating Profit Column

```python
movies ['profit'] = movies['revenue'] - movies['budget']
```

```python
movies.head(5)
```

##### 4. Removing Unwanted Columns

```python
movies_genre = movies[['popularity', 'budget', 'revenue', 'original_title', 'runtime', 'genres', 'release_year', 'vote_count', 'vote_average', 'profit']]
```

##### 5. Splitting Details in The Genres Column

```python
from pandas import Series, DataFrame
```

```python
split = movies_genre['genres'].str.split('|').apply(Series, 1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre
```

### Research Question (Q):

##### 1. Which genres are the most common (number of movies made?

```python
genre_common = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending = True)
```

```python
genre_common['original_title'].plot.barh(title = 'Movies per Genre', color = 'DarkBlue', figsize = (10, 9))
```

##### 2. Which genres have high average budget and revenue?

#####  2A. High Average Budget

```python
genre_average = movies_genre.groupby('genres_split')[['popularity', 'budget', 'revenue', 'runtime', 'vote_count', 'vote_average', 'profit']].mean()
pd.options.display.float_format = '{:2f}'.format

genre_average
```

```python
genre_average.sort_values('budget', ascending = True, inplace = True)
```

```python
genre_average[['budget', 'revenue']].plot.barh(title = 'Movies per Genre', color = ('DarkBlue', 'c'), figsize = (10, 9))
```

##### 2B. High Average Profit

```python
genre_average
```

```python
genre_average.sort_values('revenue', ascending = True, inplace = True)
```

```python
genre_average['revenue'].plot.barh(title = 'Profit by Genre', color = 'DarkBlue', figsize = (10, 9))
```

##### 3. Which genres have high average popularity?

```python
genre_average.sort_values('popularity', ascending = True, inplace = True)
```

```python
genre_average['popularity'].plot.barh(title = 'Popularity by Genre', color = 'DarkBlue', figsize = (10, 9))
```

### Research Hypotheses 

1. The best movies according to vote average high return on profit and revenue.


```python
movies.drop_duplicates(inplace = True)
movies ['profit'] = movies['revenue'] - movies['budget']
movies_genre = movies[['popularity', 'budget', 'revenue', 'original_title', 'runtime', 'genres', 'release_date', 'vote_count', 'vote_average', 'profit']]
```

```python
movies_genre.head()
```

```python
movies_counted = movies_genre[movies_genre['vote_count'] >= 50]

movies_counted[['popularity', 'budget', 'revenue', 'runtime', 'vote_count', 'vote_average', 'profit']].corr(method = 'spearman')
```

```python
sb.regplot(x = 'vote_average', y = 'profit', data = movies_counted, line_kws = {'color': 'darkgreen'})
```

###### 2. The best movies according to popularity return high profit and revenue.



```python
sb.regplot(x = 'popularity', y = 'revenue', data = movies_counted, line_kws = {'color': 'darkgreen'})
```

###### 3. Highly budgeted movies return high revenue and profit.

```python
sb.regplot(x = 'budget', y = 'profit', data = movies_counted, line_kws = {'color': 'darkgreen'})
```

###### 4. Highly budgeted movies have a high popularity.

```python
sb.regplot(x = 'budget', y = 'popularity', data = movies_counted, line_kws = {'color': 'darkgreen'})
```

```python
time_genre = pd.DataFrame(movies_genre.groupby(['release_year', 'genres_split'])['profit'].mean())
```

```python
final_genre = pd.pivot_table(time_genre, values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

###### 5. Look at Profit Per Genre Per Year

```python
sb.set(rc = {'figure.figsize': (15,10)})
sb.heatmap(final_genre, cmap = 'YlGnBu', linewidths = 0.5)
plt.title('Genres by Profit')
```
